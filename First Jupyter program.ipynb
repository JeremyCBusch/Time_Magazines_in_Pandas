```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

```

```python
df_list = pd.read_csv("timecovers.csv", encoding='latin-1')

df_list.head(20000)
```

```python
# Plot for the total occupations since the start of time magazine

df_list.drop(labels = ['Order', 'Cover #', 'POY'], axis = 1)



df_total_plot = pd.value_counts(df_list['Occupation']).plot.bar(title = "Totals for all Time Magazine Covers")
df_total_plot.set_xlabel("Number of People")
df_total_plot.set_ylabel("Occupations")
plt.show()
```

```python
#covers that were or wernt politics

df_list[df_list['Year'] > 1960].drop(labels = ['Order', 'Cover #', 'POY'], axis = 1)

df_total_plot = pd.value_counts(df_list['Occupation'] == 'Politics & Gov').plot.bar(title = "Totals for all Time Magazine Covers in Politics & Government")
df_total_plot.set_xlabel("In Politics & Government")
df_total_plot.set_ylabel("Number of People")
plt.show()
```

```python
# bar graph of the 50s

df_above_49_mask = df_list['Year'] >= 1950
df_below_60_mask = df_list['Year'] < 1960


df_list[df_above_49_mask & df_below_60_mask].drop(labels = ['Order', 'Cover #', 'POY'], axis = 1)

df_50 = pd.value_counts(df_list[df_above_49_mask & df_below_60_mask]['Occupation']).plot.bar(title = "Covers in the 50s")
df_50.set_xlabel("Occupations")
df_50.set_ylabel("Number of People")
plt.show()
```

```python
# 70s bar graph

df_above_69_mask = df_list['Year'] >= 1970
df_below_80_mask = df_list['Year'] < 1980


df_list[df_above_69_mask & df_below_80_mask].drop(labels = ['Order', 'Cover #', 'POY'], axis = 1)

df_50 = pd.value_counts(df_list[df_above_69_mask & df_below_80_mask]['Occupation']).plot.bar(title = "Covers in the 70s")
df_50.set_xlabel("Occupations")
df_50.set_ylabel("Number of People")
plt.show()
```

```python
# 20s bar graph

df_above_23_mask = df_list['Year'] >= 1923
df_below_30_mask = df_list['Year'] < 1930


df_list[df_above_23_mask & df_below_30_mask].drop(labels = ['Order', 'Cover #', 'POY'], axis = 1)

df_50 = pd.value_counts(df_list[df_above_23_mask & df_below_30_mask]['Occupation']).plot.bar(title = "Covers in the 20s")
df_50.set_xlabel("Occupations")
df_50.set_ylabel("Number of People")
plt.show()
```

```python
# 30s bar graph

df_above_30_mask = df_list['Year'] >= 1930
df_below_40_mask = df_list['Year'] < 1940


df_list[df_above_30_mask & df_below_40_mask].drop(labels = ['Order', 'Cover #', 'POY'], axis = 1)

df_50 = pd.value_counts(df_list[df_above_30_mask & df_below_40_mask]['Occupation']).plot.bar(title = "Covers in the 30s")
df_50.set_xlabel("Occupations")
df_50.set_ylabel("Number of People")
plt.show()
```

```python
politics = df_list[df_list['Occupation'] == 'Politics & Gov'].groupby('Year').size()
politics.plot(title = "Politics & Government Throught The Years")
```

```python
entertainment = df_list[df_list['Occupation'] == 'Entertainment'].groupby('Year').size()
entertainment.plot(title = "Entertainment Throught The Years")
```
